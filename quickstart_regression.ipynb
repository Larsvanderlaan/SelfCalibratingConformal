# Quickstart: Regression\n\nThis notebook shows the modernized regression workflow.

```python
import numpy as np
from selfcalibratingconformal import SelfCalibratingConformalPredictor

rng = np.random.default_rng(0)
X_cal = rng.normal(size=(80, 1))
y_cal = X_cal[:, 0] + rng.normal(scale=0.4, size=80)
X_test = rng.normal(size=(20, 1))
y_test = X_test[:, 0] + rng.normal(scale=0.4, size=20)

class Predictor:
    def predict(self, x):
        x = np.asarray(x)
        return 0.9 * x[:, 0]

model = SelfCalibratingConformalPredictor(
    Predictor(),
    algo_params={"num_bin_predictor": 16, "num_bin_y": 18},
)
model.fit(X_cal, y_cal, alpha=0.1)
coverage, width = model.check_coverage(X_test, y_test)
coverage, width
```
