# Quickstart: Venn-Abers Quantile CP\n\nThis notebook demonstrates the new quantile-loss conformal predictor.

```python
import numpy as np
from selfcalibratingconformal import VennAbersQuantileConformalPredictor

rng = np.random.default_rng(1)
X_cal = rng.normal(size=(120, 1))
y_cal = 1.25 * X_cal[:, 0] + rng.normal(scale=0.35, size=120)
X_test = rng.normal(size=(30, 1))
y_test = 1.25 * X_test[:, 0] + rng.normal(scale=0.35, size=30)

class CenterPredictor:
    def predict(self, x):
        x = np.asarray(x)
        return 1.25 * x[:, 0]

class ScoreQuantilePredictor:
    def predict(self, x):
        x = np.asarray(x)
        return 0.35 + 0.1 * np.abs(x[:, 0])

cp = VennAbersQuantileConformalPredictor(
    score_quantile_predictor=ScoreQuantilePredictor(),
    center_predictor=CenterPredictor(),
    alpha=0.1,
    algo_params={"num_bin_predictor": 18, "num_bin_score": 80},
)
cp.fit(X_cal, y_cal)
coverage, width = cp.check_coverage(X_test, y_test)
coverage, width
```
