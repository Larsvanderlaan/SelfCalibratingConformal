# Advanced customization\n\nThis notebook shows custom conformity-score hooks.

```python
import numpy as np
from selfcalibratingconformal import (
    SelfCalibratingConformalPredictor,
    VennAbersQuantileConformalPredictor,
)

rng = np.random.default_rng(3)
X_cal = rng.normal(size=(90, 1))
y_cal = 0.75 * X_cal[:, 0] + rng.normal(scale=0.3, size=90)
X_test = rng.normal(size=(10, 1))

class Predictor:
    def predict(self, x):
        x = np.asarray(x)
        return 0.7 * x[:, 0]

def regression_score(y_values, calibrated_prediction, original_predictions):
    return np.abs(y_values - calibrated_prediction) + 0.05 * np.abs(original_predictions)

reg = SelfCalibratingConformalPredictor(Predictor(), conformity_score=regression_score)
reg.fit(X_cal, y_cal, alpha=0.1)
reg.predict_interval(X_test)[:3]
```
